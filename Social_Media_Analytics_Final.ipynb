```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# The betweenness centrality of a node in a graph is the sum of the fraction of all-pairs shortest paths that pass through it.
# To calculate it, we need to create the graph from the diagram and compute the betweenness centrality for node 8.

import networkx as nx

# Create an empty graph
G = nx.Graph()

# Add edges between nodes as seen in the diagram
edges = [(1,2), (1,3), (1,4), (1,6), (2,4), (2,5), (2,7), (3,4), (3,6), (3,7), (4,5), (4,6), (4,7), (5,7), (6,7), (6,8), (6,9), (7,8), (8,9), (9,10)]

G.add_edges_from(edges)

# Calculate betweenness centrality for node 8
betweenness_centrality = nx.betweenness_centrality(G, normalized=False)
node_8_betweenness = betweenness_centrality[8]
node_8_betweenness
```
