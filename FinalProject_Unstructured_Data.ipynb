```python
!pip install pandas nltk

import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

# Download the stopwords dataset
nltk.download('stopwords')
nltk.download('punkt')

file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

descriptions = df["Description"].dropna().astype(str)

# Convert the 'descriptions' Series to a single string
text = " ".join(descriptions)

# Tokenize the string
words = word_tokenize(text)

# Convert the list of stop words to a set for faster searching
stop_words = set(stopwords.words('english'))

# Filter out the stop words and non-alphanumeric tokens from the list of words
filtered_words = [word.lower() for word in words if word.isalnum() and word.lower() not in stop_words]

# Count the occurrences of each word
word_counts = Counter(filtered_words)

# Get the most common words
most_common_words = word_counts.most_common()

print(most_common_words)
```

```python
# Extract bigrams
from nltk.util import bigrams

bigram_list = list(bigrams(filtered_words))
bigram_counts = Counter(bigram_list)

# Get the most common bigrams
most_common_bigrams = bigram_counts.most_common()

print(most_common_bigrams)

# ngram_value = 4
# ngram_list = list(ngrams(filtered_words, ngram_value))
# ngram_counts = Counter(ngram_list)
# most_common_ngrams = ngram_counts.most_common()

# print(most_common_ngrams)
```

```python
# Extract products based on user's chosen attributes
msba_attributes = ['predictive modeling', 'machine learning', 'leader']  # This is an example. This will come from user input.
def recommend_jobs_based_on_attributes(attributes_list, df):
    recommended_jobs = set()
    for attribute in attributes_list:
        subset = df[df['Description'].str.contains(attribute, case=False, na=False)]
        recommended_jobs.update(subset['Title1'].unique())

    # Return top 3 recommended jobs
    return list(recommended_jobs)[:3]

recommended_jobs = recommend_jobs_based_on_attributes(msba_attributes, df)
print(recommended_jobs)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_vectorizer = CountVectorizer(stop_words='english', lowercase=True)
count_matrix = count_vectorizer.fit_transform(descriptions)

# Convert the list of words to a numerical vector
word_vector = count_vectorizer.transform([' '.join(msba_attributes)])

# Compute cosine similarity between each review and the list of words
similarities = cosine_similarity(count_matrix, word_vector)

# Create a DataFrame to store the results
Jobs = df.dropna(subset=['Description'])
result_df = pd.DataFrame({
    'Company': Jobs['Title'],
    'Job_description': Jobs['Description'],
    'cosine_similarity': similarities.flatten()
})


# Sort by similarity score
sorted_results = result_df.sort_values(by="cosine_similarity", ascending=False)
result_df
print(sorted_results)
```

```python
title_frequencies = df['Title'].value_counts()

# Get the top 20 most frequent titles
top_20_titles = title_frequencies[:20]
```

```python
#msba_attributes
companies = top_20_titles.index.tolist()
lift_values = []

for company in companies:
    lift = 0  # Initialize lift value
    for attribute in msba_attributes:
        # Calculate Support(attribute, beer), Support(attribute), and Support(beer) here
        support_attribute_company = len(df[(df['Title'] == company) & df['Description'].str.contains(attribute)]) / len(df)
        support_attribute = len(df[df['Description'].str.contains(attribute)]) / len(df)
        support_company = len(df[df['Title'] == company]) / len(df)

        # Calculate Lift(attribute, beer)
        if support_attribute != 0 and support_company != 0:
            lift += (support_attribute_company / (support_attribute * support_company))

    lift_values.append({'Company': company, 'Lift': lift})

# Find the most similar beer based on lift values
lift_df = pd.DataFrame(lift_values)
print(lift_df.sort_values(by="Lift", ascending=False))
```
